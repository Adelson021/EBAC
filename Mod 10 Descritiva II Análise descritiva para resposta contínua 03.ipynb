````python
# Supondo que dados_clientes já esteja definido e corretamente formatado
def criar_grafico(variavel, dados_clientes):
    plt.figure(figsize=(12, 6))
    # Verifique se 'data_ref' é do tipo datetime
    if not pd.api.types.is_datetime64_any_dtype(dados_clientes['data_ref']):
        dados_clientes['data_ref'] = pd.to_datetime(dados_clientes['data_ref'], errors='coerce')
    
    # Converta 'variavel' para string se for bool
    if dados_clientes[variavel].dtype == 'bool':
        dados_clientes[variavel] = dados_clientes[variavel].astype(str)
    
    ax = sns.countplot(x=dados_clientes["data_ref"].dt.strftime('%Y-%m'), hue=variavel, data=dados_clientes)
    plt.title(f"Distribuição da variável {variavel} ao longo do tempo")
    plt.xticks(rotation=45)
    plt.xlabel("Data de referência (YYYY-MM)")
    plt.legend(title=variavel, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# Lista de variáveis qualitativas
variaveis_qualitativas = ["sexo", "posse_de_veiculo", "posse_de_imovel", "tipo_renda", "educacao", "estado_civil", "tipo_residencia", "mau"]

# Criando gráficos para cada variável
for variavel in variaveis_qualitativas:
    criar_grafico(variavel, dados_clientes)
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
def criar_grafico_medianas(variavel, dados_clientes, variavel_numerica):
    plt.figure(figsize=(12, 6))
    
    # Agrupando e calculando as médias
    dados_clientes['data_ref'] = pd.to_datetime(dados_clientes['data_ref'], errors='coerce')
    medias = dados_clientes.groupby([dados_clientes['data_ref'].dt.to_period('M').astype(str), variavel])[variavel_numerica].mean().reset_index()
    
    # Criando o gráfico
    sns.lineplot(data=medias, x='data_ref', y=variavel_numerica, hue=variavel, marker='o')
    
    plt.title(f'Médias de {variavel_numerica} por {variavel} ao longo do tempo')
    plt.xticks(rotation=45)
    plt.xlabel('Data de Referência (YYYY-MM)')
    plt.ylabel(f'Média de {variavel_numerica}')
    plt.legend(title=variavel, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# Lista de variáveis qualitativas
variaveis_qualitativas = ["sexo", "posse_de_veiculo", "posse_de_imovel", "tipo_renda", "educacao", "estado_civil", "tipo_residencia", "mau"]

# Variável numérica para calcular médias
variavel_numerica = 'renda'

# Criando gráficos para cada variável qualitativa
for variavel in variaveis_qualitativas:
    criar_grafico_medianas(variavel, dados_clientes, variavel_numerica)
````
